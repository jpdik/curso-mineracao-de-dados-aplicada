```python
import pandas as pd
import requests
import json
import time

SECRET = "&client_id=4868498968090321fc95&client_secret=b9eaa4f9707314ce7fafa88d2f947b50f00c6dc7"
```

```python
dados = []
```

```python
def get_dados(item, data):
    global dados
    
    if not 'items' in data:
        return False
    
    for i in data['items']:
        item = {}
        item['description'] = i['description']
        item['language'] = i['language']
        item['type'] = i['owner']['type']
        item['created_at'] = i['created_at']
        item['size'] = i['size']
        item['watchers_count'] = i['watchers_count']
        item['stargazers_count'] = i['stargazers_count']
        item['forks_count'] = i['forks_count']
        item['open_issues_count'] = i['open_issues_count']
        item['score'] = i['score']
        dados.append(item)
    return True
```

```python
for i in range(1, 35):
    req = requests.get("https://api.github.com/search/repositories?page="+str(i)+"&q=pushed:>=2018-11-14"+SECRET)
    time.sleep(2)
    data = json.loads(req.text)
    if not get_dados(i, data):
        break
```
