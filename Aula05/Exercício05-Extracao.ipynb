```python
def buscar_noticias(origem_url):
    dados = []
    try:
        #cada página traz 20 noticias, e foram encontradas 7877 pela API
        for i in range(1, int(7877/20)):
            all_articles = newsapi.get_everything(
                                              domains=origem_url,
                                              from_param='2018-10-26',
                                              to='2018-11-25',
                                              sort_by='relevancy',
                                              language='pt',
                                              page=i)
            for j in all_articles['articles']:
                data = j['publishedAt']
                link = j['url']
                img = j['urlToImage']
                title = j['title']
                content = j['content']
                dados.append((data, link, img, title, content))
    except Exception as e:
        print(str(e))
        return dados
    return dados
```

```python
dados = buscar_noticias('uol.com.br')
```

```python
df = pd.DataFrame(dados, columns=['data', 'link', 'img_url', 'titulo', 'descricao'])
df.tail()
```
